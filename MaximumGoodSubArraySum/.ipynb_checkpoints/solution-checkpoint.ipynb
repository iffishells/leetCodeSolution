```python
class Solution:
    def maximumSubarraySum(self, nums: List[int], k: int) -> int:
        good_sum_arry = []
        sum_list = []
        for i in range(len(nums)):
            for j in range(1,len(nums)+1):
                sub_array = nums[i:j]
                if len(sub_array)!=0:
                    
                    difference = abs(sub_array[0] - sub_array[-1]) 
    
                    if difference ==k:
                        good_sum_arry.append( sub_array)    
                        sum_list.append(sum(sub_array) )
        
        # print("good_sum_arry : ",good_sum_arry)
        # print("sum_list : ",sum_list)
        # if len(sum_list)!=0:
        # else:
        #     return 0
        return max(sum_list)
```
